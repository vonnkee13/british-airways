```python
import scrapy
from scrapy.crawler import CrawlerProcess
import pandas as pd
```

```python
url = 'https://www.airlinequality.com/airline-reviews/philippine-airlines'
```

```python
class Airline_Spider(scrapy.Spider):
    # name of the spider
    name = "air_spider"
    # define start_requests
    def start_requests(self):
        global url
        yield scrapy.Request(url = url, callback = self.parse)
    #get next link for each page then follow the next page
    def parse(self,response):
        #Get list of all tables in page
        global page
        rows = response.css('table.review-ratings') 
        for row in rows:
            if rows[0] == row:
                continue
            else:
                temp_dick = {}
                #Create dictionary of each column per table
                for column in row.xpath('tr'):
                    features = column.xpath('td//text()').extract()
                    if features[0] in ratings:
                        features[1]=len(column.xpath('td//span[@class="star fill"]').extract())
                    temp_dick[features[0]]=features[1]
                table.append(temp_dick)
        page = page + 1
        if page > 343:
            break
        yield response.follow(url = 'https://www.airlinequality.com/airline-reviews/philippine-airlines/page'+str(page)+"/",callback = self.parse)
            
table = []
page = 1
            
ratings = ['Seat Comfort','Cabin Staff Service','Food & Beverages','Inflight Entertainment','Ground Service','Value For Money', 'Wifi & Connectivity']   
```

```python
process = CrawlerProcess() # initialize Crawler
process.crawl(Airline_Spider) # initialize JS Spider
process.start()

```

```python
df = pd.DataFrame(table)
```
